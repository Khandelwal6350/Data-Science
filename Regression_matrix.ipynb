```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
```

```python
data = {
    "cgpa": [6.89, 5.12, 7.82, 7.42, 6.94,
             8.10, 7.15, 5.90, 8.50, 6.50],
    
    "package": [3.26, 1.98, 3.25, 3.67, 3.57,
                4.20, 3.80, 2.40, 4.80, 3.10]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
a=df.loc[1, 'cgpa']
```

```python
print(a)
```

```python
plt.scatter(df['cgpa' ], df[ 'package' ])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]

X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python

lr.fit(X_test,y_test)
```

```python
plt.scatter(df["cgpa"],df["package"])
plt.plot(X_train,lr.predict(X_train),color="red")
plt.xlabel("cgpa")
plt.ylabel("package in lap")
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,accuracy_score
```

```python
y_pred= lr.predict(X_test)
```

```python
lr.predict(X_test)
```

```python
y_test.values
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
```

```python
print("MAE",mean_squared_error(y_test,y_pred))
```

```python
print("MAE",np.sqrt(mean_absolute_error(y_test,y_pred)))
```

```python
print("MAE",r2_score(y_test,y_pred))
```
