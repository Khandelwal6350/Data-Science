```python
"""
__________________________Covariance___________________________________

In Machine Learning, Covariance measures how two variables change together.

Covariance is a statistical measure that indicates the direction of the relationship 
between two variables. It shows whether two variables increase or decrease together.


Types of Covariance
____________________________1. Positive Covariance_______________________

When two variables move in the same direction.

Example:

Height ↑ → Weight ↑
Temperature ↑ → Ice cream sales ↑

Meaning:

Both variables increase or decrease together.

____________________________2. Negative Covariance________________________

When two variables move in opposite directions.

Example:

Speed ↑ → Travel time ↓
Product price ↑ → Demand ↓

Meaning:

One variable increases while the other decreases.

___________________________3. Zero Covariance_____________________________

When there is no relationship between variables.

Example:

Shoe size and exam marks

Meaning:

Variables are independent or unrelated.

	​
"""





import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("tips_dataset_10x7.csv")
```

```python
df.info()
```

```python
df.select_dtypes(["float","int"])
```

```python
df["total_bill"].sum()
```

```python
cor=df.select_dtypes(["float","int64"]).corr()
```

```python
cov=df.select_dtypes(["float","int64"]).cov()
```

```python
plt.Figure(figsize=(4,3))
sns.heatmap(cor)
plt.show()
```

```python
plt.Figure(figsize=(4,3))
sns.heatmap(cov)
plt.show()
```
